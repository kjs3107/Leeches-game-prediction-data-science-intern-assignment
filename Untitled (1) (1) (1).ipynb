```python
import matplotlib.pyplot as plt

# Assuming you have imported chess_df_cleaned already

# Calculate average moves for white and black players
avg_white_turn = data[data['winner']=='white']['turns'].mean()
avg_black_turn = data[data['winner']=='black']['turns'].mean()

# Calculate win rates for white and black players
total_white_matches = len(data[data['winner']=='white'])
total_black_matches = len(data[data['winner']=='black'])
total_matches = len(data)
win_rate_white = total_white_matches / total_matches
win_rate_black = total_black_matches / total_matches

# Plotting
labels = ['Average Moves', 'Win Rate']
white_values = [avg_white_turn, win_rate_white]
black_values = [avg_black_turn, win_rate_black]

x = range(len(labels))

plt.bar(x, white_values, width=0.4, label='White', color='lightblue')
plt.bar([i + 0.4 for i in x], black_values, width=0.4, label='Black', color='lightgreen')

plt.xlabel('Metrics')
plt.ylabel('Values')
plt.title('Comparison of White vs. Black Players')
plt.xticks([i + 0.2 for i in x], labels)
plt.legend()
plt.show()
```

<h2>IMPUTING MISSING OUTLIERS</h2>
</div>

```python
from sklearn.impute import KNNImputer
import numpy as np

def remove_outliers_and_impute(data):
    # Define numerical columns
    numerical_columns = ['turns', 'white_rating', 'black_rating', 'opening_ply', 'total_second']

    # Remove outliers using IQR
    Q1 = data[numerical_columns].quantile(0.25)
    Q3 = data[numerical_columns].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    cleaned_data = data[((data[numerical_columns] > lower_bound) & (data[numerical_columns] < upper_bound)).all(axis=1)]

    # Impute missing values using KNN Imputer
    imputer = KNNImputer(n_neighbors=5)
    imputed_data = imputer.fit_transform(cleaned_data[numerical_columns])
    cleaned_data[numerical_columns] = imputed_data

    return cleaned_data
```
